## Arreglos en Python
Son listas de datos en los que hay un orden, por lo que es importante tener en cuenta a la posición del elemento. El primer elemento es el número 0, y no el número 1 (Cosas de Python). En las listas se pueden modificar sus elementos, y puede haber elementos duplicados. Se crean poniendo sus elementos entre corchetes [a,b,c, >>>].
Aquí algunas funciones para modificar los arreglos y algunos ejemplos:
- ```lista[i]```: Devuelve el elemento que está en la posición ```i```.
- ```lista.pop(i)```: Devuelve el elemento en la posición ```i``` de una lista y luego lo borra.
- ```lista.append(elemento)```: Añade elemento al final de la lista.
- ```lista.insert(i, elemento)```: Inserta elemento en la posición ```i```.
- ```lista.extend(lista2)```: Fusiona ```lista``` con ```lista2```.
- ```lista.remove(elemento)```: Elimina la primera vez que aparece elemento.

````python
a = [10, 20, 30, 40, 50]
````

## if 
Existen algunas situaciones en la programación donde es necesario implementar una condición, para algunos valores y otra condición para otros. 
Esto se logra con el comando ``` if ``` para las condiciones principales, y ```elif``` para condiciones adicionales, o ```else``` si es que no se cumplen las condiciones anteriores.
FB_IMG_1601785487467.jpg

Las condiciones que se suelen usar con más frecuencia son:
- ```a == b``` –> Indica si a es igual a b
- ```a < b```
- ```a > b```
- ```not``` –> NO: niega la condición que le sigue.
- ```and``` –> Y: junta dos condiciones que tienen que cumplirse las dos
- ```or``` –> O: junta dos condiciones y tienen que cumplirse alguna de las dos.
En el siguiente ejemplo, puede cambiar los valores condicionales para tener los distintos mensajes programados.

````python
lista_1 =[1,2,3,4,5,6,7,8] 
````

````python
len(lista_1)
````

````python
for ele in range(len(lista_1)):
  if lista_1[ele] == 3:
    print(lista_1[ele], "Es correcto")
  elif lista_1[ele] < 3:
    print("Hola")
  elif lista_1[ele] > 3:
    print("Adios")
  else:
    print("No es un numero")
````

````python
for ele in range(len(lista_1)):
  if lista_1[ele] == 3:
    print(lista_1[ele], "Es correcto")
````

## while
En ocasiones, tenemos que repetir varias veces una determinada tarea hasta conseguir nuestro objetivo. En Python esto se realiza con el comando while. A modo de ejemplo while en Python se usa así:

````python
cnt = 0
while cnt < 10:
  print(cnt, "adios")
  cnt += 1
````
