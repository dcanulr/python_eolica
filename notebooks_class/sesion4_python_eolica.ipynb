## Lanzar 50 monedas 10 veces

```
experimentos = []

for i in range(50): #numero de experimentos
  tiros = []

  for i in range(10): #numero de lanzamientos en cada experimentos
    tiros.append(random.randint(0,1))

  experimentos.append(sum(tiros)/10)
```

```
# experimentos
```

## Histogramas

```
import matplotlib.pyplot as plt
```

```
plt.plot(experimentos)
```

```
plt.hist(experimentos, bins=20, cumulative=True, density=True) #probabilidades
```

```
plt.hist(experimentos, bins=20, density=True) #frecuencias
```

```
 def monedas(exp, lanz): 
  experimentos = []

  for i in range(exp): #numero de experimentos
    tiros = []

    for i in range(lanz): #numero de lanzamientos en cada experimentos
      tiros.append(random.randint(0,1))

    experimentos.append(sum(tiros)/lanz)

  print(sum(experimentos)/len(experimentos))  
  plt.hist(experimentos, bins=20)
```
